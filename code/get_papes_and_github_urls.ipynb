```python
import pandas as pd
import glob
import json
import github_api
import search_for_url
import numpy as np
```

```python
papes = glob.glob('../git_papes/*')
```

```python
papes = sorted(papes)
```

```python
len(papes)
```

```python
with open(papes[5]) as infile:
    dat = json.load(infile)
```

```python
pairs = []
for p in papes:
    with open(p) as infile:
        dat = json.load(infile)
        for d in dat:
            #assert(len(d['documents'])==1)
            pmid = d['documents'][0]['id']
            urls = search_for_url.search_for_url(' '.join([i['text'] for i in d['documents'][0]['passages']]))
            for u in urls:
                if 'github' in u:
                    pairs.append((pmid, u))
pairs = np.array(pairs)
```

```python
urls_in_papes = pd.DataFrame({'pmid': np.array(pairs)[:, 0], 'url': np.array(pairs)[:, 1]})
```
