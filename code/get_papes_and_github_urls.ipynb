````python
def standardize_url(u):
    pp = parse.urlparse(u)
    u = pp.netloc + pp.path
    if u[-1]=='/':
        u = u[:-1]
    return u
````

````python
pairs = []
for p in papes:
    with open(p) as infile:
        dat = json.load(infile)
        for d in dat:
            #assert(len(d['documents'])==1)
            pmcid = d['documents'][0]['id']
            urls = search_for_url.search_for_url(' '.join([i['text'] for i in d['documents'][0]['passages']]))
            for u in urls:
                u = standardize_url(u)
                if 'github' in u:
                    pairs.append((pmcid, u))
pairs = np.array(pairs)
````

## Positive feature heuristic

````python
# check if abstract section or title 
positive_feats = ['code', 'custom', 'develop', 'documentation', 
                  'download', 'host', 'license',
       'manual', 'publish', 'raw', 'reproduc', 'script', 'source', 'update', 'present']
positive_feats = positive_feats + [f.capitalize() for f in positive_feats]
````

````python
hit_table = {}

for pmcid in np.unique(urls_in_papes.pmcid.values):
    textt = ' '.join([t['text'] for t in pmcid_to_doc[pmcid]['passages'] if 'github' in t['text']])
    hits_vec = [(f in textt) for f in positive_feats]
    hit_table[pmcid] = np.sum(hits_vec)
````

````python
hit_table = pd.DataFrame.from_dict(hit_table, orient = 'index', columns = ['score'])
````

````python
no_hits = hit_table.index[hit_table.score == 0]
has_hits = hit_table.index[hit_table.score > 0]
````
